```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from matplotlib import pyplot as plt
import cv2
import math
from tqdm import tqdm
import os
```

```python
image = cv2.cvtColor(plt.imread("friends.jpeg"), cv2.COLOR_RGB2GRAY)/255.
#apply edge detection using Sobel
sobelx = cv2.Sobel(image,cv2.CV_64F,1,0,ksize=21)
sobely = cv2.Sobel(image,cv2.CV_64F,0,1,ksize=21)
magnitude = np.sqrt(np.power(sobelx,2)+ np.power(sobely,2))
phase = np.empty_like(sobely)
for i in range(sobely.shape[0]):
    for j in range(sobely.shape[1]):
        if not sobelx[i,j]==0:
            phase[i][j] = math.atan(sobely[i,j]/sobelx[i,j])
        else: 
            phase[i][j] = math.atan(sobely[i,j]/math.inf)#if the phase is 90o
```

```python
plt.axis(False)
plt.imshow((magnitude), cmap="gray")
```

```python
plt.imshow((phase), cmap="gray")
```
